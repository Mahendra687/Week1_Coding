**KDE plot**
* K: kernal
* D: density
* E: estimation
# 

* KDE, allow us t estimate **probability** density function from finite dataset

**Univariate:KDE plot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import*
from pandas import*
```

```python
sns.set()
```

**Datasets**

```python
cars = sns.load_dataset("mpg").dropna()

print("Shape of the data", cars.shape)
print("Dimension of the data",cars.ndim)
print("Data types in data",cars.dtypes)
```

```python
cars.head(3)
```

```python
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower, shade=True);
```

```python
sns.kdeplot(cars.horsepower, bw=10);
# small the bw = (bandwidth) increases the variance and higher the bw will loses the modes
```

```python
sns.kdeplot(cars.horsepower, cumulative=True);
```

**Bivariate: KDEplot**
* Can estimat the joint probability desnsity function in 2D

```python
sns.kdeplot(cars.horsepower, cars.mpg);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels = 20);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, shade=True);
```

```python
sns.set_style("whitegrid")

sns.kdeplot(cars.horsepower, cars.mpg, shade=True, shade_lowest=False);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, 
             shade=True, shade_lowest=False, cbar=True);
```
