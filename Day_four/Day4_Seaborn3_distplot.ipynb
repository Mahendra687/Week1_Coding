**distplot**
* Distribution plot is called distplot
* There are many way to plot distribution of dataset: histogram, KDEplot, and rugplot
* Seaborn combine these plot feature in one to see distribution in more precise way

**Import Libraries**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import*
from numpy import*
```

**Load the datsets**

```python
cars = sns.load_dataset("mpg").dropna()

print("Data types:",cars.dtypes)
print("Shape of the Datasets:", cars.shape)
print("Dimension of the datasets:", cars.ndim)
```

```python
cars.head(3)
```

**Plotting the dataset**

```python
sns.set_style('white')
sns.displot(cars.weight, kde = True); # Histogram + KDEplot = distplot
```

```python
sns.displot(cars.weight, kde=False);
```

```python
#sns.displot(cars.weight, hist=False);
```

```python
sns.displot(cars.weight,rug=True); # hist + rugplot
```

**Bin**

```python
sns.displot(cars.weight, bins = 20);
```

```python
sns.displot(cars.weight, bins = 10);
```

```python
sns.displot(cars.weight, bins=[1500,2000,2500,3500,5200]);
```
