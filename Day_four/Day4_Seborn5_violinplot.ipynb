**Violinplot** = boxplot+KDEplot
* Similar to the boxplot
* kind of symetric plot
* it alert to show us kernal density estimation

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import*
from numpy import*
```

```python
cars = sns.load_dataset('mpg').dropna()

print("The data types:",cars.dtypes)
print("The shape of the data:", cars.shape)
print("The dimension of the data:", cars.ndim)
```

```python
cars.head(3)
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])] 
```

```python
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(x = cars.cylinders, y=cars.displacement);
```

```python
sns.violinplot(x ='cylinders',y='displacement', data = cars);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
              data=cars);
```

**Violinplot Options**

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe','usa'])])
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])])
plt.legend(loc=2);
```

**Split**

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])], split=True)
plt.legend(loc=2);
```

**inner**

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])], split=True, inner='quartiles')
plt.legend(loc=2);
```

**Scale**

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])], split=True,scale='count')
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])], split=True,scale='count',
               scale_hue=False)
plt.legend(loc=2);
```

```python
(
    cars[cars.origin.isin(['japan','europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

**Styling**

```python
sns.violinplot(x=cars.origin,y=cars.displacement, 
               order=['japan','europe','usa' ]);
```

```python
sns.violinplot(x=cars.origin,y=cars.displacement,linewidth=3);
```

**KDE options**

```python
sns.violinplot(c=cars.cylinders,y=cars.displacement, bw=0.2);
```
