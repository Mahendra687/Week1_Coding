**Scatterplot**
* The scatter-plot consider as relational plot
* Goal is see the relationship btwn two variables

```python
import seaborn as sns
from matplotlib import pyplot as plt 
```

```python
diamonds = sns.load_dataset('diamonds')

print(diamonds.dtypes)
```

```python
print(type(diamonds))
```

```python
print(diamonds.shape)
```

```python
print(diamonds.ndim)
```

```python
diamonds.describe()
```

```python
diamonds.describe(include=['category'])
```

```python
diamonds.head(3)
```

```python
diamonds = diamonds[diamonds.cut.isin(['Premium','Good']) &
                    diamonds.color.isin(['D','F','J'])
                    ].sample(n=100, random_state=22) # will sampling down 100 observation
diamonds.shape
```

```python
sns.scatterplot(data=diamonds, x='carat', y='price')
plt.xlabel='carat'
plt.ylabel='price'
plt.grid()
plt.show()
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds);
```

```python
sns.scatterplot(x='x', y='y', data=diamonds); # x and y measurement for each diamonds
```

**hue**

```python
sns.set_style('darkgrid')

sns.scatterplot(x='carat', y='price', hue='cut',data=diamonds);
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',data=diamonds, 
                palette="tab10"); # palette wil specify the color
```

```python
sns.scatterplot(x='carat',y='price',hue='depth',data=diamonds); # will give the depth of continuous variables
```

**Styple**

```python
sns.scatterplot(x='carat',y='price',data=diamonds, s=100,style='cut');
```

**size**

```python
sns.scatterplot(x='carat',y='price',size='cut',data=diamonds);
```

```python
sns.scatterplot(x='carat',y='price',size='cut',data=diamonds,
               sizes=[200,150,100,50,25]); # for categories
```

```python
sns.scatterplot(x='carat',y='price',size='depth',data=diamonds); # for continuous
```

**hue+style+size**

```python
sns.scatterplot(x='carat',y='price',hue='cut',data=diamonds,s=100);
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',style='cut'
                ,data=diamonds,s=100);
```

```python

sns.scatterplot(x='carat',y='price',hue='cut',style='color'
                ,data=diamonds,s=100);
```

**alpha**

```python
sns.scatterplot(x='carat',y='price'
                ,data=diamonds,s=100,alpha=0.6); # transparancy
```

**marker styling**

```python
sns.scatterplot(x='carat',y='price'
                ,data=diamonds,s=200,marker='*');
```

```python
sns.scatterplot(x='carat',y='price'
                ,data=diamonds,s=200,marker='*', edgecolor='blue');
```

```python
sns.scatterplot(x='carat',y='price'
                ,data=diamonds,s=200,marker='*', edgecolor='green');
```

```python
sns.scatterplot(x='carat',y='price'
                ,data=diamonds,s=200,marker='*', edgecolor='red');
```
