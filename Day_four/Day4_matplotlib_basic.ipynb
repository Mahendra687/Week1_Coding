3. **Change figure size, plot red dots,set axis scales x: 0-6 and y: 0-20**

```python
plt.figure(figsize=(2,5)) # x and y values in inches: 2 inches wide x 5 inches tall
plt.plot([1,2,3,4], [1,4,9,16], 'ro') # r: color red and o: shape: rp,rx, kx
plt.axis([0, 6, 0, 20]) # [xmin, xmax, ymin, ymax] scaling the axis: 0-6 on the x-axis and 0-20 on the y-axis
plt.annotate("square it", (3,6))
plt.show()
```

6. **Time series - six months of random floats**

```python
ts = pd.Series(np.random.randn(180), index=pd.date_range("1/1/2018", periods=180))
df = pd.DataFrame(np.random.randn(180,3), index=ts.index, columns=list("ABC"))
df.cumsum().plot()
plt.show()     
```

7. **Random dots in a Scatter**

```python
N = 50
x = np.random.rand(N) # x- axis
y = np.random.rand(N) # y- axis
colors = np.random.rand(N) # 50 random colors
sizes = (30*np.random.rand(N)**2) # 0-15 point radii
plt.scatter(x, y, s = sizes, c = colors, alpha=0.5) # alpha is transparacny setting
plt.show()
```

9. **Subplot, part1**
# 
* 221 = top left subplot
* 222 = top right subplot
* 223 = bottom left subplot
* 224 = bottom right subplot

```python
fig  = plt.figure()
fig.suptitle("My_Subplot")
fig.add_subplot(221) # add plot1
plt.plot([np.log(n) for n in range(1,10)]) # for log-curve(line)
fig.add_subplot(222, facecolor = 'y')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()
```

9.2 **Subplots:Part2**

```python
fig, plots = plt.subplots(2, sharex = True)
fig.suptitle("Sharing X-axis")
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x, y, color = "r")
plots[1].scatter(x, y)
plt.show()
```
